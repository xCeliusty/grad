```python
import os
import pandas as pd
import numpy as np
import sys

excel_src = r"C:\Users\amals\Desktop\Trafic prediction\Dataset"
excel_dest = r"C:\Users\amals\Desktop\Trafic prediction\Merged\Dataset.xlsx"
excels = [i for i in os.listdir(excel_src) if i.split('.')[-1] == 'xlsx']
#print(excels)

# Reading Calendar Dataset
calendar_loc = excel_src + "\\" + excels[0]
calendar = pd.read_excel(calendar_loc, index_col=None, keep_default_na=False, usecols="A,B")
#print(calendar)

# Reading Traffic Dataset
traffic_loc = excel_src + "\\" + excels[1]
traffic = pd.read_excel(traffic_loc, index_col=None, keep_default_na=False, usecols="A,B,C")
#print(traffic)

# Reading Weather Dataset
weather_loc = excel_src + "\\" + excels[2]
weather = pd.read_excel(weather_loc, index_col=None, usecols="A,B,C,E,N", keep_default_na=False)
#print(weather)

# Create Dictionary of Zones
zones = {}
count = 1
for i in weather['name'].unique():
  zones[i] = count
  count = count + 1
#print(zones)

# Create Dictionary of CodedDay
days = {'Saturday':1, 'Sunday':2, 'Monday':3, 'Tuesday':4, 'Wednesday':5, 'Thursday':6, 'Friday':7}
#print(days)

# Create Format of Merged Data
columns = {'Day': [],
        'Date': [],
        'CodedDay': [],
        'Zone': [],
        'Weather': [],
        'Temperature': [],
        'Rain': [],
        'Holiday': [],
        'Traffic': []}
data = pd.DataFrame(columns)
#print(data)

# Creating columns of Data
dateCol = []
dayCol = []
codedDayCol = []
zoneCol = []
weatherCol = []
tempCol = []
rainCol = []
holidayCol = []
trafficCol = []
for i in range(weather['name'].count()):
  dateCol.append(weather['datetime'][i].strftime("%m/%d/%Y"))
  dayCol.append(weather['datetime'][i].day_name())
  codedDayCol.append(days[weather['datetime'][i].day_name()])
  zoneCol.append(zones[weather['name'][i]])
  weatherCol.append(weather['tempmax'][i])
  tempCol.append(weather['temp'][i])
  rainCol.append(1 if weather['preciptype'][i] else 0)
  holidayCol.append(int(calendar.loc[calendar['Date'] == dateCol[i]]['Holiday']))
  trafficCol.append(int(traffic.loc[traffic['Date']== dateCol[i]].loc[traffic['Zone']== weather['name'][i]]['Traffic']))

# Creating Data as Panda Data-Frame
data['Date'] = dateCol
data['Day'] = dayCol
data['CodedDay'] = codedDayCol
data['Zone'] = zoneCol
data['Weather'] = weatherCol
data['Temperature'] = tempCol
data['Rain'] = rainCol
data['Holiday'] = holidayCol
data['Traffic'] = trafficCol
#print(data)

data = data.sort_values(['Date', 'Zone'], ascending=[True, True])
#print(data)
data.to_excel(excel_dest, index=False)
```
